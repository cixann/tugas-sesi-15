import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

bulan = ['Januari','Februari','Maret','April','Mei','Juni','Juli','Agustus','September','Oktober','November','Desember']

#Pengeluaran Antara 500.000 - 2.000.000

minggu_1 = np.random.randint(500000,2000001,12)
minggu_2 = np.random.randint(500000,2000001,12)
minggu_3 = np.random.randint(500000,2000001,12)
minggu_4 = np.random.randint(500000,2000001,12)

data = {
    'bulan' : bulan,
    'minggu_1' : minggu_1,
    'minggu_2' : minggu_2,
    'minggu_3' : minggu_3,
    'minggu_4' : minggu_4,
}

df = pd.DataFrame(data)

df['total pengeluaran'] = df[['minggu_1','minggu_2','minggu_3','minggu_4']].sum(axis = 1)
df

```python
#total peneluaran perbulan

plt.figure (figsize=(12,6))#canvas dengan ukuran width,height(inc)
plt.title('Total Pengeluaran Per Bulan')
plt.bar(df['bulan'],df['total pengeluaran'],color =['skyblue'])
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.ylabel('Rupiah')
plt.tight_layout()
plt.show()
```

```python
#Scatter plot untuk analisis data pengeluaran per minggu per bulan
plt.figure(figsize=(12,6)) 
for minggu in ['minggu_1','minggu_2','minggu_3','minggu_4']:
    plt.plot(df['bulan'],df[minggu],marker='o',label=minggu)

plt.title("pengeluaran per Minggu")
plt.xlabel('bulan')
plt.ylabel('pengeluaran (Rupiah)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#pie chart -> prasentase
plt.figure(figsize=(6,6))
plt.pie(df['total pengeluaran'],labels=df['bulan'],startangle=90,autopct='%1.1f%%')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

df['bulan_anggka'] = np.arange(1,13)

x = df[['bulan_anggka']]
y = df['total pengeluaran']

model = LinearRegression()

model.fit(x,y)

prediksi = model.predict([[13]])[0]
print(f"hasil prediksi adalah:{prediksi}")
```
